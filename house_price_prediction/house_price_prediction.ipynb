# OutLier Detection and Removal

```python
dataset.describe()
```

```python
# 1 room whose square feet is 300 is not feasibel
dataset = dataset[(dataset["total_sqft"]/dataset["bhk"]) >= 300]
(dataset["total_sqft"] / dataset["bhk"]).describe()
```

```python
# values between the mean-SD and mean+SD is consider
def remove_outliers_per_sqft(df):
  dataset_output = pd.DataFrame()
  for key , subdf in dataset.groupby("location"):
    m = np.mean(subdf.Price_per_sqft)
    st = np.std(subdf.Price_per_sqft)

    gen_df = subdf[(subdf.Price_per_sqft > (m-st)) & (subdf.Price_per_sqft <= (m+st))]
    dataset_output = pd.concat([dataset_output,gen_df],ignore_index=True)
  return dataset_output

data = remove_outliers_per_sqft(dataset)
data.describe()
```

```python
def bhk_outlier_remover(df):
  exclude_indices = np.array([])
  for location , location_df in df.groupby("location"):
    bhk_stats = {}
    for bhk , bhk_df in location_df.groupby("bhk"):
      bhk_stats[bhk] = {
          "mean" : np.mean(bhk_df.Price_per_sqft),
          "std" : np.std(bhk_df.Price_per_sqft),
          "count" : bhk_df.shape[0]
      }


      for bhk , bhk_df in location_df.groupby("bhk"):
        stats = bhk_stats.get(bhk-1)
        if stats and stats["count"] > 5:
          exclude_indices = np.append(exclude_indices , bhk_df[bhk_df.Price_per_sqft < (stats["mean"])].index.values)
  return df.drop(exclude_indices , axis="index")
```

```python
data = bhk_outlier_remover(data)
```

```python
X = data.drop(columns=["price"])
y = data["price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Lasso , Ridge
from sklearn.preprocessing import OneHotEncoder , StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Apply Linear Regression

```python
column_trans = make_column_transformer(
    (OneHotEncoder(sparse=False), ["location"]),  # OneHotEncoder for 'location'
    remainder="passthrough"  # Pass through other columns without transformation
)
```

```python
scaler = StandardScaler()
```

```python
model = LinearRegression()
```

```python
pipe = make_pipeline(column_trans , scaler , model)
```

```python
pipe.fit(X_train , y_train)
```

```python
y_pred_model = pipe.predict(X_test)
```

```python
r2_score(y_test , y_pred_model)
```

# Applying Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans , scaler , lasso)
```

```python
pipe.fit(X_train , y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
r2_score(y_test , y_pred_model)
```

# Applying Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(
    column_trans,  # Column transformer for preprocessing
    StandardScaler(),  # Scaler for numerical features
    Ridge()  # Ridge regression model
)
```

```python
pipe.fit(X_train , y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
r2_score(y_test , y_pred_model)
```

```python
print("No Regularization" , r2_score(y_test , y_pred_model))
print("Lasso" , r2_score(y_test , y_pred_lasso))
print("Ridge" , r2_score(y_test , y_pred_ridge))
```
